```python
import flash
from flash.audio import SpeechRecognition, SpeechRecognitionData
from flash.core.data.utils import download_data
import torch
torch.cuda.empty_cache()
```

```python
!unzip /content/Archivio.zip -d /content/data
#download_data("https://pl-flash-data.s3.amazonaws.com/timit_data.zip", "./data")
```

```python
datamodule = SpeechRecognitionData.from_json(
    input_fields="file",
    target_fields="text",
    train_file="/content/data/train.json",
    test_file="/content/data/test.json",
)
```

```python
model = SpeechRecognition(backbone="facebook/wav2vec2-base-960h")
model.model.freeze_feature_extractor()
model.to('cuda')
```

```python
trainer = flash.Trainer(gpus=1,max_epochs=50, checkpoint_callback=True)
trainer.finetune(model, datamodule=datamodule, strategy='no_freeze')
```

```python
predictions = model.predict(["/content/data/TEST/19_4.wav"])
print(predictions)
```

```python
model.to('cpu')
# Save Checkpoint 
trainer.save_checkpoint("./speech_recognition_model.pt")
# Load Model From Checkpoint 
#model = SpeechRecognition.load_from_checkpoint("./speech_recognition_model.pt")
```
